```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculate total number of schools
total_schools = school_data_complete['school_name'].nunique()
print(total_schools)
```

```python
#Calculate the total number of students
total_students = school_data_complete['student_name'].count()
print(total_students)
```

```python
#Calculate the total budget(use school_data to not count duplicates on combined data)
total_budget = school_data['budget'].sum()
print(total_budget)
```

```python
#Calculate the average math score
average_math_score = school_data_complete['math_score'].mean()
print(round(average_math_score, 6))
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete['reading_score'].mean()
print(round(average_reading_score, 5))
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = student_data.loc[student_data["math_score"] >= 70]
passing_math_count = passing_math["student_name"].count()
percentage_students_passing_math = ((passing_math_count/total_students)*100)
print(round(percentage_students_passing_math, 6))
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = student_data.loc[student_data["reading_score"] >= 70]
passing_reading_count = passing_reading["student_name"].count()
percentage_students_passing_reading = ((passing_reading_count/total_students)*100)
print(round(percentage_students_passing_reading, 6))
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]['student_name'].count()/total_students)*100
print(round(overall_passing,6))
```

```python
#Create a dataframe to hold the above results
district_summary= pd.DataFrame({
    "Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": (round(average_math_score, 6)),
    "Average Reading Score": (round(average_reading_score, 5)),
    "% Passing Math": (round(percentage_students_passing_math, 6)),
    "% Passing Reading": (round(percentage_students_passing_reading, 6)),
    "% Overall Passing Rate": (round(overall_passing,6))
}, index =[0])
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type

  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math

* Create a dataframe to hold the above results

```python
#summary table by student showing passing math and reading
school_data_complete["passing_math"] = school_data_complete["math_score"] >= 70
school_data_complete["passing_reading"] = school_data_complete["reading_score"] >= 70

school_data_complete
```

```python
#groupby school
school_names = school_data_complete.groupby(["school_name"]).mean()
#formatting passing_math
school_names['passing_math'] = round(school_names['passing_math']*100,6)
#formatting passing_reading
school_names['passing_reading'] = round(school_names['passing_reading']*100,6)
#calculate overall passing rate
school_names['overall_passing'] = round(((school_names['passing_math'] * school_names['passing_reading']/100)),6)
#school_names
```

```python
#adding school type
school_types = school_data.set_index("school_name")["type"]
#school_types

#merge to include all data
school_summary = pd.merge(school_types, school_names, how = "left", on=["school_name"])
school_summary = school_summary.sort_values(["school_name"], ascending = True)
school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#top 5 performing schools
top_5_schools = school_summary.sort_values(["overall_passing"], ascending = False)
top_5_schools.head()
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_5_schools = school_summary.sort_values(["overall_passing"], ascending = True)
bottom_5_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade and groupby school

ninth_grade=  school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_grade =  school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade =  school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_grade=  school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

#Combine the series into a dataframe
math_scores = pd.DataFrame({"9th Grade":ninth_grade, "10th Grade":tenth_grade, 
                                     "11th Grade":eleventh_grade, "12th Grade":twelfth_grade})
math_scores
```
